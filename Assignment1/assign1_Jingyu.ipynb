```python
# Question 11
from keras import models, layers
```

```python
# Question 12
myNetwork = models.Sequential()
```

```python
# Question 13
train_images = train_images.reshape((60000, 28 * 28))
test_images = test_images.reshape((10000, 28 * 28))

train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# Question 14
myNetwork.add(layers.Dense(512, activation='relu', input_shape=(784,)))
```

```python
# Question 15
myNetwork.add(layers.Dense(128, activation='relu'))
```

```python
# Question 16
myNetwork.add(layers.Dense(10, activation='softmax'))
```

```python
# Option A
myNetwork.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_A = myNetwork.fit(
    train_images,
    train_labels,
    epochs=10,
    validation_split=0.2
)
```

```python
# Option B
from keras.utils import to_categorical

train_labels_cat = to_categorical(train_labels)
test_labels_cat = to_categorical(test_labels)

myNetwork.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_B = myNetwork.fit(
    train_images,
    train_labels_cat,
    epochs=10,
    validation_split=0.2
)
```

```python
# Evaluation
test_loss, test_acc = myNetwork.evaluate(test_images, test_labels_cat)
print("Test accuracy:", test_acc)
print("Test loss:", test_loss)
```
